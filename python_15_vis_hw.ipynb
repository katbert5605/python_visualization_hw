```python
df_happy_15['year'] = 2015
df_happy_16['year'] = 2016
```

### Взаимосвязь величин
Рассмотрим взаимосвязь очков уровня счастья от разных составляющих на точечных диаграммах

```python
df_happy.plot(kind='scatter',
             x='Happiness Score',
             y='Generosity',
             title='Корреляция между очками счастья и щедростью')
df_happy.plot(kind='scatter',
             x='Happiness Score',
             y='Freedom',
             title='Корреляция между очками счастья и свободой')
plt.show()
```

Четкой взаимосвязи между щедростью и очками счастья не прослеживается. 
Взаимосвязь между осками счастья и свободой прослеживается лучше. Можем сделать вывод, что в странах, где высокий уровень счастья не встречается очень низких оценок свободы.

```python
# Посмотрим эти три составляющие в пузырях
df_happy.plot(kind='scatter',
             x='Freedom',
             y='Generosity',
#              Для того, чтобы лучше визуализировать разницу, возведем очки счатья в какую-либо степень
             s=df_happy['Happiness Score']**3,
             title='Корреляция между свободой, щедростью и очками счастья')
plt.show()
```

Общая зависимость видна лучше, чем на отдельных точечных гистграммах. 
Вместе с тем, можно обратить внимание на города с относительно высоким уровенем очков счастья и одновремено низкими оценками по свободе и щедрости. 

```python
# Посмотрим на среднюю величину очков счастья в разрезе регионов и закрасим показатели прямо в ячейках 
df_happy.groupby(['Region'])[['Happiness Score']].mean()\
    .sort_values('Happiness Score', ascending=False).style.bar(align='mid', color='green')
```

### jointplot
Посмотрим **зависимость очков счатья** от других покзаателей

```python
columns=['Economy (GDP per Capita)', 'Family', 'Trust (Government Corruption)', 'Health (Life Expectancy)']
for column in columns:
    sns.jointplot(x='Happiness Score', y=column, data=df_happy, kind='reg')
    plt.show()
```

На основании графиков можем сделать выводы, что кол-во очков счастья в большей степени зависит от экономики, семьи и здооровья. Уровень доверия оказывает влияние, но слабее.

## И закончим *тепловой картой*


```python
#изменим размер визуализации 
#plt.figure(figsize=(20,10)
#построим сводную таблицу зависимости очков счастья по странам (columns) и округленному уровню экономики
econ_round = df_happy['Economy (GDP per Capita)'].round(1)
hp_score = df_happy.pivot_table(index=econ_round,
                     columns='Region',
                     values='Happiness Score',
                     aggfunc='mean').sort_index(ascending=False)
hp_score
```

```python
# Получили много NaN значений. Посмотрим, как на них отреагирует heatmap 
plt.figure(figsize=(20,10))
sns.heatmap(hp_score)
```

**Балгодаря тепловой карте** видим **влияние экономики** на кол-во очков счастья. Вместе с тем, стоит обратить влияние на регионы пустыни Сахара и юговосточной Азии. Вне зависимости от оценки уровня экономики, оценка счатья в данно регионе низкая.
